```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_selection import VarianceThreshold
from sklearn.feature_selection import mutual_info_classif
from sklearn.decomposition import PCA
from sklearn.manifold import LocallyLinearEmbedding, Isomap, TSNE
```

```python
#Leemos el CSV de entrenamiento
file_path = "./traintabs/"
file_name = "traintab01.csv"
X_init = pd.read_csv(file_path+file_name, sep = ',', decimal = '.', index_col=0)
#Quitamos la columna de´nombre de la foto que no nos da ninguna info
X_subinit = X_init.iloc[:, 1:]
X_subinit.head()
```

```python
#Escalamos a intervalo unidad
unit_scaler = MinMaxScaler().set_output(transform="pandas")
unit_scaler.fit(X_subinit)
X_scl = unit_scaler.transform(X_subinit)
X_scl.head()
```

```python
#Vamos a aplicar filtros para quitar dimensionalidad a nuestro dataframe
#Filtro de varianza en el dataframe normalizado

#Threshold de 0.01, lo que tenga menor varianza que este valor se eliminará
var_th = 0.01

feat_selector = VarianceThreshold(var_th).set_output(transform='pandas')
feat_selector.fit(X_scl)
X_sel = feat_selector.transform(X_scl)

#Mostramos que cambios se han hecho
strMsg = X_sel.columns.to_list()
print('%d Características borradas'  % (X_subinit.columns.shape[0]-len(strMsg)) ) 
print('%d Características seleccionadas:'% len(strMsg))
print( strMsg )
```

```python
#Ahora vamos a aplicar un filtro de correlación
#Veamos si existe una dependencia lineal entre las columnas de nuestro dataframe
mat_R = X_sel.corr()
R_th = .995

aux = np.abs( np.triu(mat_R.values) - np.eye(101) )
ind_row = np.argmax(aux, axis=0)
ind_col=0
aboveTh_list=[]
removed_list=[]
for ir in ind_row:
    if (aux[ir,ind_col] >= R_th):
        aboveTh_list.append((ir,ind_col))
        removed_list.append(ind_col)
    ind_col=ind_col+1

print('Pares con correlación superior a %0.4f son:'%R_th)
[print(' ',X_sel.columns[aux[0]],'-',X_sel.columns[aux[1]]) for aux in aboveTh_list]
print('Las columnas eliminadas son:')
drop_list = [aux for aux in X_sel.columns[removed_list] ]
print(' ',drop_list)

#Eliminamos las columnas
X_sel2 = X_sel.drop(columns=drop_list)
X_sel2.head()
```
