# Non-linear classification

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_circles, make_moons

n_samples = 300
X,y = make_circles(n_samples = n_samples, noise=0.2, factor=0.5, random_state=1)

plt.plot(X[y==0][:,0],X[y==0][:,1],'o')
plt.plot(X[y==1][:,0],X[y==1][:,1],'rx')

X3 = X[:,0]**2 + X[:,1]**2
X3 = np.expand_dims(X3, axis=1)
X  = np.concatenate((X,X3), axis=1)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(X[y==0][:,0], X[y==0][:,1], X[y==0][:,2] , marker='o')
ax.scatter(X[y==1][:,0], X[y==1][:,1], X[y==1][:,2] , marker='x')
```
